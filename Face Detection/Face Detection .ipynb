
### Analyze the OpenCV function detectMultiScale and the database of face images with respect to automating a “sweep” of parameter values as well as template matching to find the images of US President Obama.
### Function that lists all of the contents of a directory and determines:  
### 1) Where the faces are. 
### 2) Using your own designed template matching, which faces are President Obama.
### 3) The result of your function should generate: list of image names of the directory that have Obama’s face and cropped bounding box images of the images.

```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import re
import glob
import os
import sys

# Define process_images_from_directory.
# This function will import the directories and then pass them into the program.

def process_images_from_directory(faceCascade, directory, scaleFactor=1.1, minNeighbors=8):
    imageDictionary = dict() 
    for directoryPath, directoryNames, fileNames in os.walk(directory):                                       # walk path 
        for fileName in fileNames:
            imageFile = os.path.join(directoryPath, fileName)                                                 # Join path            
            img = cv2.imread(imageFile)                                                                       # Reads image into RGB for histogram comparison
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                                                    # RGB option (not used)
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # GRAY option (used)     
            # Scale using faceCascade.detectMultiScale - haarcascade_frontalface_default.xml
            faces = faceCascade.detectMultiScale(img_gray, scaleFactor=scaleFactor, minNeighbors=minNeighbors, minSize=(101,101), maxSize=(800,800))
            # Get rect format for later use 
            for (x, y, w, h) in faces:
                image = cv2.rectangle(img_gray,(x,y),(x+w,y+h),(255,255,255),10)                              # NOTE: Putting into white space for easy manipulation
                roi_color = image[y:y + h, x:x + w]                                                           # ROI for better manipulation within program calls                        
                imageDictionary[fileName] = (img_gray, roi_color, img, imageFile)                             # Push all needed vars into dictionary
    return imageDictionary

# Define match_images
# This function does the work of inputting images and processing the images for exporting. 
# This function also prints the found Obamas with the problem statements requirements.

def match_images(dict_in, template):  
    for image in dict_in:         
        results1 = []
        results2 = []
        count1 = 0
        count2 = 0
        res1 = cv2.matchTemplate(dict_in[image][1],template['ObamaTemplate1.jpg'][1],cv2.TM_CCOEFF_NORMED)   # Test images against matchTemplate   
        res2 = cv2.matchTemplate(dict_in[image][1],template['ObamaTemplate2.jpg'][1],cv2.TM_CCOEFF_NORMED)   # Test images against matchTemplate  
    #print(A_dict[image][2])
        for r in res1:                                                                                       # First for loop for counting matches
            threshold = .8
            if (r >= threshold).any:
                count1 += 1
            results1=count1
        for r in res2:                                                                                       # Second for loop for counting matches
            threshold = .8
            if (r >= threshold).any:
                count2 += 1
            results2=count2
        if results1<375 and results2>100:                                                                    # Use KMEANS to draw line for result match 
            print("A wild Obama has been found in folder",dict_in[image][3],"\n", "\n")
            plt.figure()
            img_gray = cv2.cvtColor(dict_in[image][0], cv2.COLOR_BGR2RGB)
            plt.imshow(img_gray)
            plt.title(dict_in[image][3])
            plt.show()
            print("\n", "\n")
            return dict_in[image][0]
        
# Enter loop for processing. 
# The below code will enter a path of haar cascade frontal face default and define a path.

OPENCV_PATH = "/usr/local/lib/python3.6/dist-packages/cv2/data"
faceXML = os.path.join(OPENCV_PATH,'haarcascade_frontalface_default.xml')
faceCascade = cv2.CascadeClassifier(faceXML)
baseDir = "/home/CV/FaceData"
#/home/CV/FaceData
#/home/CV/FaceData_Test

# Enter import the directories and then pass them into the program.

A_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'A'))
B_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'B'))
C_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'C'))
D_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'D'))
E_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'E'))
ObamaTemplate_dict = process_images_from_directory(faceCascade,os.path.join(baseDir,'ObamaTemplate'))

# Enter inputting images and processing the images for exporting.
# This task also prints the found Obamas with the problem statements requirements.

print("We are now looping through folder A which has the contents of:","\n")
print(A_dict.keys(),"\n")
match_images(A_dict,ObamaTemplate_dict)
print("We are now looping through folder B which has the contents of:","\n")
print(B_dict.keys(),"\n")
b = match_images(B_dict,ObamaTemplate_dict)
print("We are now looping through folder C which has the contents of:","\n")
print(C_dict.keys(),"\n")
c = match_images(C_dict,ObamaTemplate_dict)
print("We are now looping through folder D which has the contents of:","\n")
print(D_dict.keys(),"\n")
d = match_images(D_dict,ObamaTemplate_dict)
print("We are now looping through folder E which has the contents of:","\n")
print(E_dict.keys(),"\n")
e = match_images(E_dict,ObamaTemplate_dict)

# End 
```
